Import the package and create an instance of the Turtle:

````python
from pjnturtle.turtle import Turtle
t = Turtle()
````

### 1. The drawing area
The drawing area has a size of 1024 $\times$ 1024 pixels by default. The pixel in the upper left corner has coordinates ```(x = 0, y = 0)``` while the lower left has ```(x = 0, y = 1024)``` . The drawing area is initialized on creating a Turtle with ```t = Turtle()```. Two commands are important:

* ```t.clear()``` emties the current drawing area.
* ```t.home()``` positions the Turtle in the middle of the drawing area and directs it to the north.


### 2. A first drawing
Turtle graphics can be thought of as a Turtle holding a pen and moving around. We can control the Turtle with several commands to draw a picture:

* ```t.penDown()``` puts the pen down.
* ```t.penUp()``` puts the pen up.


* ```t.forward(x)``` moves the Turtle forward by ```x``` steps.
* ```t.backward(y)``` moves the Turtle backward by ```y``` steps.


* ```t.turnLeft(x)``` turns the Turtle to the right by ```x``` degrees.
* ```t.turnRight(y)``` turns the Turtle to the left by ```y``` degrees.


* ```t.drawTurtle()``` draws a Turtle at the current position with the current direction.
* ```t.display()``` displays the drawing.

Now we can draw our first Turtle graphic.

````python
t.clear()
t.home()
t.penDown()
t.forward(600)
t.turnRight(180)
t.penUp()
t.forward(600)
t.penDown()
t.forward(600)
t.penUp()
t.backward(600)
t.turnRight(90)
t.forward(600)
t.turnLeft(180)
t.penDown()
t.forward(1200)
t.home()
t.drawTurtle()
t.display()
````

### 3. Colors and pen size
To change the color of the lines and the size following commands are available:

* ```t.penSize(x)``` sets the pen size to ```x``` pixel.
* ```t.penColor('BLUE')``` changes the color of the pen to ```'BLUE'```.
* ```t.backgroundColor('BLACK')``` enables you to change the background color to ```'BLACK'```. This has to be done before ```t.clear()``` to have an effect.

Available colors are ```'BLACK'```, ```'BLUE'```, ```'GREEN'```, ```'GREY'```, ```'ORANGE'```, ```'PURPLE'```,```'RED'```,```'WHITE'``` and ```'YELLOW'```.

* ```t.color(r,g,b)```  enables you to use your own color given by ```r```, ```g``` and ```b``` values between 0 and 255 together with ```t.penColor(t.color(255, 102, 153)```.
* ```t.addColor('MYCOLOR',t.color(r,g,b))``` gives the possiblity to add a named color that can be used with the same Turtle instance again. However, a new instance ```Turtle()``` provides only the default colors.
* ```t.reset()``` sets the pen size and color to the default values 1 and ```'BLACK'```.

Here an example of those commands in action:

````python
t.addColor('LIGHTGREY', t.color(215, 215, 215))
t.backgroundColor('LIGHTGREY')
t.clear()
t.home()
t.penSize(5)
t.penColor(t.color(255, 102, 153))
t.forward(200)
t.turnRight(120)
t.penSize(1)
t.penColor('YELLOW')
t.addColor('LIME', t.color(212, 255, 0))
t.forward(200)
t.turnRight(120)
t.penColor('LIME')
t.penSize(10)
t.forward(200)
t.reset()
t.forward(200)
t.display()
````

### 4. Additional commands
To change the color of the lines and the size following commands are available:

* ```t.goto(x, y)``` places the turtle to coordinates ```x,y```.
* ```t.setOrientation(0)``` changes the turtles orientation to the desired angle in degrees. ```0``` is north, ```90``` is to the east, ```180``` is south and ```270``` is to the west.

* ```t.write(text, fontSize)``` inserts a text box with the provided ```text``` and ```fontSize```. The text color is the same as the current pen color.
* ```t.fill()``` fills the current background area enclosed by lines with current pen color..

Again some examples:

````python
t.backgroundColor('WHITE')
t.clear()
t.reset()
t.goto(338, 712)
t.setOrientation(60)
t.penDown()
t.penColor('BLUE')
t.forward(400)
t.setOrientation(-60)
t.forward(400)
t.setOrientation(180)
t.forward(400)
t.penUp()
t.turnLeft(160)
t.forward(200) # inside the white area, enclosed by the triangle
t.penColor('YELLOW')
t.fill()

t.drawTurtle()
t.penColor('BLACK')
t.write('Here am I', 16)
t.display()
````

### 5. Some fun
The methods from PJNturtle can be used in functions and other methods as well. Below you find an example with recursion resulting in a spiral hexagon.

````python
def hexagon(l, a, s):
    colVec = ['PURPLE','BLUE','GREEN','YELLOW','ORANGE','RED']
    i = 0
    while True:
        if  (-2 < l < 2) and (-2 < l < 2):
            break
        t.penColor(colVec[i])
        t.forward(l)
        t.turnRight(a)
        l = l-s
        if i == 5:
            i = 0
        else:
            i = i + 1

t = Turtle()
t.backgroundColor('BLACK')
t.clear()
t.goto(512,312)
t.penDown()
t.setOrientation(120)
hexagon(200, 60.8, 0.5)
t.display()
````
